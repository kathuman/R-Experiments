```r
library(qcc)
```

```r
diameters <- as.data.frame(replicate(4, rnorm(10,mean=1.31,sd=0.05)))
```

```r
diameters
```

```r
q <- qcc(diameters, type="R", nsigmas=3)
```

```r
q <- qcc(diameters, type="xbar", nsigmas=3)
```

```r
process.capability(q, spec.limits=c(1.31,1.32))
```
